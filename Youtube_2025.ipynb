```python
import pandas as pd

try:
#reading csv file
    data_set = pd.read_csv('Downloads/youtube_2025_dataset_202504061614.csv')
    data_set.head()
    
except FileNotFoundError as e :
    print(e)
```

```python
#checking for null values
missing_values = data_set.isnull().sum()
print('Missing values in each column:')
print(missing_values)
```

```python
data_set.info()
```

```python
data_set.describe()
```

```python
#handling missing values
data_set = data_set.fillna('Unknown')
```

```python
data_set.head()
```

```python
#checking for duplicates
duplicates = data_set[data_set.duplicated()]
print(duplicates)

#dropping duplicates
data_set.drop_duplicates()
```

```python

data_set.columns = data_set.columns.str.replace(' ', '_')
data_set.columns = data_set.columns.str.replace('(', '')
data_set.columns = data_set.columns.str.replace(')', '')
data_set.columns = data_set.columns.str.replace('%', '')
```

```python
#connecting to MYSQL
import mysql.connector

conn = mysql.connector.connect(
    host = "localhost",
    user = "root",
    password = "#Tharun123#",
    database = "Youtube_data"
)
cursor = conn.cursor(buffered = True)
```

```python
#creating MYSQL engine
from sqlalchemy import create_engine

engine = create_engine("mysql+mysqlconnector://root:#Tharun123#@localhost/Youtube_data")
print(data_set.columns)
print(data_set.dtypes)
```

```python
#storing of values in database
data_set.to_sql('data_set', con=engine, if_exists='replace', index=False)
```

```python
# Total number of videos
cursor.execute("SELECT COUNT(Total_Videos) AS Total_Videos FROM data_set")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Total_Videos'])
print(df)
```

```python
# Most engaging channels assuming Engagement_Score
cursor.execute("""
    SELECT Channel_Name, SUM(Engagement_Score) AS Total_Engagement
    FROM data_set
    GROUP BY Channel_Name
    ORDER BY Total_Engagement DESC
    LIMIT 10
""")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Channel_Name', 'Total_Engagement'])
print(df)
```

```python
# Get the top 5 channels with the most subscribers
cursor.execute("""
    SELECT Channel_Name, Youtuber_Name, Total_Subscribers
    FROM data_set
    ORDER BY Total_Subscribers DESC
    LIMIT 5
""")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Channel_Name', 'Youtuber_Name', 'Total_Subscribers'])
print(df)
```

```python
#Count how many channels use AI-generated content
cursor.execute("""
    SELECT COUNT(*) AS Channels_Using_AI_Content
    FROM data_set
    WHERE AI_Generated_Content_ = 'Yes'
""")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Channels_Using_AI_Content'])
print(df)
```

```python
# Getting YouTubers and their channels that support neural interfaces 
cursor.execute("""
    SELECT Channel_Name, Youtuber_Name
    FROM data_set
    WHERE Neural_Interface_Compatible = 'Yes' AND Metaverse_Integration_Level != 'None'
""")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Channel_Name', 'Youtuber_Name'])
print(df)
```

```python
#to know unique vales from a given column
cursor.execute("SELECT DISTINCT Metaverse_Integration_Level FROM data_set")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Metaverse_Integration_Level'])
print(df)
```

```python
# Retrieving the count of entries grouped by each level of metaverse integration
cursor.execute("""
    SELECT Metaverse_Integration_Level, COUNT(*) AS Count
    FROM data_set
    GROUP BY Metaverse_Integration_Level
""")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Metaverse_Integration_Level', 'Count'])
print(df)
```

```python
# average Engagement_Score for each Metaverse_Integration_Level
cursor.execute("""
    SELECT Metaverse_Integration_Level, AVG(Engagement_Score) AS Avg_Engagement_Score
    FROM data_set
    GROUP BY Metaverse_Integration_Level
""")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Metaverse_Integration_Level', 'Avg_Engagement_Score'])
print(df)
```

```python
#channels that have Neural_Interface_Compatible set to True
cursor.execute("""
    SELECT Channel_Name, Youtuber_Name
    FROM data_set
    WHERE Neural_Interface_Compatible = 'True'
""")
res = cursor.fetchall()
df = pd.DataFrame(res, columns=['Channel_Name', 'Youtuber_Name'])
print(df)
```
